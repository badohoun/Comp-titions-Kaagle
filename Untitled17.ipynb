Fonction de preprocessing pour adapter les données à nos algorithmes 


1.   Codage de variable catégorielle 
2.   Concatenation avec des variables numériques en un seul dataframe 



1.   conversion en float bien interprétés par les algos de ML
2.   Nouvelle colonne label : Yes = 1 , No = 0  







```
def preprocessing(df):


  df_cat = df[['City' , 'Gender']]

  df_cat = pd.get_dummies(df_cat)
  df_num = df[['Age' , 'Income' , 'Illness']]

  df = pd.concat([df_cat , df_num] , axis = 1)

  df['Illness'] = np.where(df['Illness'] == 'No' , 0  , 1 )

  return df



toy_dataset = preprocessing(toy_dataset)

toy_dataset.head()
```



1. Features et Target 
2.Séparation en jeu de données d'apprentissage et de test (train_test)



```
features = toy_dataset.iloc[:, : 12  ]
target = toy_dataset.iloc[:,12:]

train_features , test_features ,  train_labels , test_labels = train_test_split(features , target , test_size = 0.25 ,  random_state = 42)
```



1.   Modèle  : Réseaux de neurones simple (modèle séquentiel)



La construction d'un réseau de neurones nécéssite d'empiler des couches de neurones appélés dense parce que les données d'entrées sont connectés aux sorties de manière dense , par ailleurs on appelle l'ensemble du modèle de réseau neuronal  modèle séquentiel car il nécessite d'enchainer ces couches appelés denses de manière séquentiels . On applique donc sur les couches intermédiaires ou cachées  une fonction d'activation qui à pour role d'introduire de la non linéarité dans les données afin de captez les structures complexes qui existent dans nos données 

Ensuite on compile le modèle , ici trois éléments fondamentaux :



*   La perte (loss) mesure le cout d'une mauvaise prédiction
Il existe différentes types de fonctions de pertes qui sont classés en 2 catégories :
Les fonctions de perte sont associés soit aux problématiques de classification soit aux problématiques de régression 

- Régression : Mean squared Error 
- Classification : binary_cross_entropy 

Ceci est évidemment une liste non exhaustive des fonctions de pertes 






*  La métrique pour surveiller ce qui se passe lors de l'étape d'entrainement  et de test .



Enfin on l'optimizeur son role est de minimizer la fonction de perte en son poids , en clair , la perte étant fonction de poids , on cherche le poids w qui minimise la fonction de perte , c'est le role de l'optimizeur  .



Ensuite on fait qu dropout pour évitez que notre modèle fasse de l'underfitting : c'est a dire le fait que notre modèle n'apprenne pas tout des données ou de l'overfitting c'est à dire le fait que notre modèle est trop complexe et donc se généralise mal aux données qui n'a jamais vu auparavant . 


En effet le dropout est une technique de régularisation qui permet de supprimer aléatoirement les activations sur les couches selon un ratio ou un pourcentage que l'on lui spécifie généralement la moitié afin que notre modèle ne fasse pas de lien entre les différents noeuds et qu'il se force à trouver les données manquantes .


Une deuxième technique de régularisation existe et s'appelle earling stopping : 
L'objectif ici est de visualisez l'étape d'entrainement et de test sur une courbe et de s'arreter juste avant que notre modèle overfit sur la courbe 





Après ce topo , construisons notre réseau de neurones avec une couche de 64 neurones comme dimension en entrée on lui spécifie le nombre de variables qui sera introduit dans le réseau : 12, la fonction d'activation pour introduire la non linéarité ici on choisit : Relu(x) = max(0,x)

Ensuite une couche de dropout à 50% , ensuite une autre couche dense de 64 neurones avec toujours comme activation Relu(x) = max(0,x)
Une couche dropout à 50% et enfin une couche dense en sortie avec un neurone et comme fonction d'activation la fonction sigmoid répérsentant la fonction de répartition de la loi logistique . 


```
def build_model():

  model = tf.keras.Sequential([tf.keras.layers.Dense(units = 64 , input_dim = 12 , activation = 'relu'),
                               tf.keras.layers.Dropout(rate = 0.5),
                               tf.keras.layers.Dense(units = 64 , activation = 'relu'),
                               tf.keras.layers.Dropout(rate = 0.5),
                               tf.keras.layers.Dense(units = 1 , activation = 'sigmoid')
                               ])
  
  model.compile(loss = 'binary_crossentropy',
                metrics = ['accuracy'],
                optimizer = 'rmsprop')
  


  return model 
```

```
%load_ext tensorboard
```

```
model1 = build_model()

log_dir = "logs/fit" + datetime.datetime.now().strftime("%Y%M%D-%H%M%S")

tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir = log_dir , histogram_freq = 1)

model1.fit(train_features , 
          train_labels , 
          epochs = 20, 
          batch_size = 128 , 
          callbacks = [tensorboard_callback]
          )
```

Ici on ne lui transmet pas toutes les données à notre réseau immédiatement on passe par lots , d'ou le batch_size dans notre modèle à 128 , ensuite le nombre d'epochs qui lui précise combien de fois on lui transmet les lots à travers le réseau tout en ajustant les poids à chaque fois .

On peut aussi utiliser les memes données que précedemment mais le transformez en tf.data.dataset pour benéficier de la capacité de traitements par lots 



```
train_dataset = tf.data.Dataset.from_tensor_slices((train_features , train_labels))

test_dataset = tf.data.Dataset.from_tensor_slices((test_features , test_labels))



train_dataset = train_dataset.shuffle(6000).batch(64)

test_dataset = test_dataset.batch(64)
```

```
model2 = build_model()

log_dir = "logs/fit" + datetime.datetime.now().strftime("%Y%M%D-%H%M%S")

tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir = log_dir , histogram_freq = 1)

model2.fit(train_dataset ,  
          epochs = 20, 
          callbacks = [tensorboard_callback]
)
```
