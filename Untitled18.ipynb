# Statistiques descriptives en amont de notre mod√®le 

Dans ce cadre, il faut effectuer une description de chaque variable quantitative et qualitative s√©paremment car les √©tapes ne sont pas les memes .

Voici la r√©partition des variables et les √©tapes √† r√©aliser pour chaque type de variable :

**Quantitative : sepal_length	sepal_width	petal_length	petal_width**

Moyenne, √©cart-type, m√©diane, minimum et maximum .

Histogramme pour avoir un aper√ßu de la distribution, et √©ventuellement comparer √† une gaussienne
Test de normalit√© √† ajouter 
Bo√Æte √† moustaches pour estimer la pr√©sence ou non d'outliers



**Qualitative** : species


Table d'effectifs ,  pourcentages et diagramme en barres 

Notons que le diagramme circulaire est √† √©viter quand on a trop de modalit√©s , ici ce n'est pas le cas .

Nous allons faire une fonction permettant la description d'une variable (en fonction de son type bien √©videmment), et l'utiliser pour chacune des colonnes du jeu de donn√©es.

```
def descQuanti(var):
    # description d'une variable quantitative du jeu de donn√©es iris
    # r√©cup√©ration des valeurs
    x = iris_df[var]
    # affichage des diff√©rentes informations demand√©es
    print("Variable   :", var, "\n")
    print("Moyenne    : ", round(np.mean(x)))
    print("Ecart-type : ", round(np.std(x)))
    print("M√©diane    : ", round(np.median(x)))
    print("Minimum    : ", round(np.min(x)))
    print("Maximum    : ", round(np.max(x)))
    # cr√©ation d'un graphique avec 2 zones
    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
    # cr√©ation de l'histogramme avec la densit√©
    h = x.plot(kind = "hist", density = True, color = "lightgrey", ax = ax1)
    x.plot(kind = "kde", ax = ax1)
    # on y ajoute la p-value du test de normalit√©
    t = sc.normaltest(x)
    plt.text(.9, .9, "Normalit√© (p-val) : " + str(t.pvalue),
                           horizontalalignment='right',
                           verticalalignment='top',
                           transform = h.transAxes)
    # cr√©ation de la bo√Æte √† moustache
    iris_df.boxplot(column = var, grid = False, ax = ax2)

# on applique la fonction sur la variable 'sepal_length' 
descQuanti("sepal_length")
# √† faire sur les autres variables
# descQuanti("petal_width")
# ...
```

```
def descQuali(var):
    # description d'une variable qualitative du jeu de donn√©es iris 
    # r√©cup√©ration des valeurs
    x = iris_df[var]
    # affichage des diff√©rentes informations demand√©es
    print("Variable   :", var, "\n")
    # tableau des effectifs et des proportions
    eff = pd.crosstab(x, "Effectif")
    pro = pd.crosstab(x, "Pourcentage", normalize=True) * 100
    print(pd.concat([eff, pro.round(2)], axis = 1))
    # diagramme en barres
    eff.plot(kind = "bar", legend = False)
    eff.plot.pie(subplots = True , figsize = (6,6))

# on lance sur la variable "cut"
descQuali("species")
# √† faire sur les autres variables
# descQuali("color")
# ...
```

# Lien entre une variable quantivative et une autre (quantitative ou qualitative)

```
#Cr√©ation d'une fonction affichant le nuage de points (avec la corr√©lation)
def lienSepal_lengthQuanti(var):
    # nuage de points entre 	sepal_length et la variable pass√©e en param√®tre pour le jeu de donn√©es iris 
    n = iris_df.plot.scatter(var, "sepal_length")
    # calcul de la corr√©lation
    cor = iris_df.sepal_length.corr(iris_df[var])
    # affichage de la corr√©lation
    plt.text(.9, .1, "Corr√©lation : " + str(cor.round(2)),
                           horizontalalignment='right',
                           verticalalignment='top',
                           transform = n.transAxes)

# on lance sur la variable "sep"
lienSepal_lengthQuanti("petal_length")
# √† faire sur les autres variables
# lienPriceQuanti("sepal_width")
# ...
```

```
# Cr√©ation d'une fonction pour √©tablir le lien entre sepal_length et une variable quali
def lienSepal_lengthQuali(var):
    # on r√©cup√®re les valeurs de la variable
    x = iris_df[var]
    # cr√©ation des bo√Ætes √† moustaches
    b = iris_df.boxplot(column = "sepal_length", by = var, grid = False) 
    # ajout d'un espace en haut du graphique pour y mettre la p-value de l'ANOVA
    b.set_ylim(0, np.max(iris_df.sepal_length) * 1.25)
    # calcul de l'ANOVA
    g = [iris_df.sepal_length[x == s] for s in list(x.unique())]
    aov = sc.f_oneway(*g)
    # ajout de la p-value dans le graphique
    plt.text(.5, .9, "ANOVA (p-value) : " + str(aov.pvalue.round(5)),
                           horizontalalignment='center',
                           verticalalignment='top',
                           transform = b.transAxes)

# on lance sur la variable "cut"
lienSepal_lengthQuali("species")
# √† faire sur les autres variables
# lienPriceQuali("color")
# ...
```
